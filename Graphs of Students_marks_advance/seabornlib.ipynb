```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)
marks = np.random.randint(30, 100, size=100)

df = pd.DataFrame({
    "Student_ID": range(1, 101),
    "Gender": np.random.choice(["Male", "Female"], size=100),
    "Subject": np.random.choice(["Math", "Science", "English"], size=100),
    "Marks": marks,
    "Attendance": np.random.randint(50, 100, size=100)
})

display(df.head()) 
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Marks"], kde=True, bins=15, color="indigo")
plt.title("Distribution of Student Marks")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Attendance", y="Marks", hue="Gender")
plt.title("Marks vs Attendance")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Subject", y="Marks", hue="Subject", palette="Set2", legend=False)
plt.title("Marks Distribution by Subject")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["Marks","Attendance"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x="Gender", y="Attendance", color="green", inner="quartile")
plt.title("Attendance Distribution by Gender")
plt.show()
```

```python
sns.regplot(data=df, x="Attendance", y="Marks")
```

```python
sns.pairplot(df[["Marks", "Attendance", "Gender"]], hue="Gender", diag_kind="hist")
plt.show()
```
